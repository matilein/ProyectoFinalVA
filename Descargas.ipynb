```python
# ==============================
# üîΩ Descargar todos los vuelos TBBR (Flug1_100 ‚Üí Flug1_105)
#     con sus anotaciones COCO
# ==============================

# Crear carpeta para el dataset
mkdir -p TBBR && cd TBBR

# Instalar zstd si no est√°
sudo apt-get update -y
sudo apt-get install -y zstd

# Lista de vuelos disponibles
for ID in 100 101 102 103 104 105; do
    echo "‚¨áÔ∏è Descargando Flug1_${ID}..."
    wget -c https://zenodo.org/records/7022736/files/Flug1_${ID}.tar.zst -O Flug1_${ID}.tar.zst

    echo "üì¶ Descomprimiendo Flug1_${ID}..."
    tar -I zstd -xvf Flug1_${ID}.tar.zst -C .

    echo "‚úÖ Flug1_${ID} listo."
    echo "------------------------------------"
done

# Descargar archivos de anotaciones COCO (uno por vuelo)
for ID in 100 101 102 103 104 105; do
    echo "‚¨áÔ∏è Descargando anotaciones Flug1_${ID}..."
    wget -c https://zenodo.org/records/7022736/files/Flug1_${ID}Media_coco.json -O Flug1_${ID}Media_coco.json
done

# Mostrar contenido final
echo "‚úÖ Descarga completa. Archivos en:"
ls -lh
```

```
import numpy as np
import matplotlib.pyplot as plt

# Cargamos la imagen (B, G, R, Thermal, Height)
img = np.load('train/images/Flug1_100/DJI_0979_R.npy')
print("Shape:", img.shape)

# Separar canales
b = img[:, :, 0]
g = img[:, :, 1]
r = img[:, :, 2]
thermal = img[:, :, 3]
height = img[:, :, 4]

# Construir imagen RGB (invertimos canales a RGB y normalizamos)
rgb = np.stack([r, g, b], axis=2) / 255.0

# Mostrar todos los canales
plt.figure(figsize=(18,10))

plt.subplot(2,3,1)
plt.imshow(rgb)
plt.title("RGB (color real)")
plt.axis("off")

plt.subplot(2,3,2)
plt.imshow(thermal, cmap='inferno')
plt.title("Térmico")
plt.axis("off")

plt.subplot(2,3,3)
plt.imshow(height, cmap='terrain')
plt.title("Mapa de altura")
plt.axis("off")

plt.tight_layout()
plt.show()
```
